```python
from sqlalchemy import Column, Integer, String, Date, Text, VARCHAR, Float , MetaData, Table , select, create_engine
from sqlalchemy import ForeignKey, Sequence,func
from sqlalchemy.ext.declarative import declarative_base
import pandas as pd 

Base = declarative_base()

db_string = "postgres://postgres:postgres@localhost:5432/EPL"
engine = create_engine(db_string)
```

```python
db = pd.read_csv('EPL_20_21.csv')
db.head(10)
```

```python
db.info()
```

```python
class Statistics (Base):
    __tablename__ = 'statistics'
    id = Column(Integer, primary_key=True)
    matches = Column(Integer)
    starts = Column(Integer)
    min_played = Column(Integer)
    goals = Column(Integer)
    assists = Column(Integer)
    passes_attempted = Column(Integer)
    perc_passes_completed = Column(Float)
    penalty_goals = Column(Integer)
    penalty_attempted = Column(Integer)
    yellow_cards = Column(Integer)
    red_cards = Column(Integer)
    def __repr__(self):
        return "<statistics(id='{0}', matches='{1}', starts='{2}', min_played='{3}', goals='{4}',assists='{5}',\
    passes_attempted='{6}', %_passes_completed='{7}', penalty_goals='{8}', penalty_attempted='{9}',yellow_cards='{10}',\
    red_cards='{11}')>".format(self.id, self.matches, self.starts, self.min_played, self.goals, self.assists,\
    self.passes_attempted, self.perc_passes_completed, self.penalty_goals, self.penalty_attempted, self.yellow_cards,\
    self.red_cards)
```

```python
class Players (Base):
    __tablename__ ='players'
    id = Column(Integer, primary_key=True)
    name = Column(String(225), nullable=False)
    club = Column(Integer,ForeignKey('clubs.id'))
    nationality = Column(Integer,ForeignKey('countries.id'))
    position = Column(String(50))
    age = Column(Integer, nullable=False)
    statistics_id = Column(Integer,ForeignKey('statistics.id'))
    def __repr__(self):
        return "<players(id='{0}', name='{1}', club='{2}', nationality='{3}', position='{4}', age='{5}',statistics_id='{6}')>"\
    .format(self.id, self.name, self.club, self.nationality, self.position, self.age, self.statisctics_id)
```

```python
class Countries (Base):
    __tablename__= 'countries'
    id = Column(Integer, Sequence('seq_country_id'), primary_key=True)
    country_name=Column(String(50), nullable= False)
    def __repr__(self):
              return "<countries(id='{0}'), country_name='{1}'>".format(self.id, self.country_name)
```

```python
class Clubs (Base):
    __tablename__ = 'clubs'
    id = Column(Integer, primary_key=True)
    name = Column(String(50),nullable=False)
    def __repr__(self):
        return "<clubs(id='{0}', name='{1}')>".format(self.id, self.name)
```

```python
Base.metadata.create_all(engine)
```

```python
def select_players(club):
    '''
    Selecting club members by club number
    :param club: int
    :return df: DataFrame, ['player_id'] int, ['player_name'] str
    :return club_name: str
    '''
    mapper_stmt = select([dic_table['players'].c.id, dic_table['players'].columns.name]).\
        where(dic_table['players'].c.club == club).order_by(dic_table['players'].c.name)
    mapper_results = engine.execute(mapper_stmt).fetchall()
    df = pd.DataFrame(mapper_results, columns=('player_id','player_name'))
    mapper_stmt = select([dic_table['clubs'].columns.id, dic_table['clubs'].columns.name])
    club_names = engine.execute(mapper_stmt).fetchall()
    club_name = club_names[club][1]
    return df, club_name
```

## EPLs teams average age

```python
mapper_stmt = select([dic_table['players'].c.club, dic_table['clubs'].c.name, func.avg(dic_table['players'].columns.age)]).\
            select_from(dic_table['clubs'].join(dic_table['players'], dic_table['clubs'].c.id == dic_table['players'].c.club)).\
            group_by(dic_table['players'].c.club, dic_table['clubs'].c.name).order_by(func.avg(dic_table['players'].columns.age))
mapper_results = engine.execute(mapper_stmt).fetchall()
avg_age = pd.DataFrame(mapper_results, columns=['club_id','club_name','avg_age'])
avg_age['avg_age'] = avg_age['avg_age'].astype(float).round(2)
avg_age.style.set_caption("Average age of EPLs teams")
```

## Top 10 most scoring players in EPL

```python
mapper_stmt = select([dic_table['players'].c.id, dic_table['players'].c.name, dic_table['statistics'].c.goals]).\
    select_from(dic_table['players'].join(dic_table['statistics'], dic_table['players'].c.statistics_id == \
    dic_table['statistics'].c.id)).order_by(dic_table['statistics'].c.goals.desc()).limit(10)
mapper_results = engine.execute(mapper_stmt).fetchall()
most_scoring_players = pd.DataFrame(mapper_results, columns=('player_id','player_name','goals'))
most_scoring_players
```

## Top 20 players that spent the most time on field

```python
mapper_stmt = select([dic_table['players'].c.id, dic_table['players'].c.name, dic_table['statistics'].c.min_played]).\
    select_from(dic_table['players'].join(dic_table['statistics'], dic_table['players'].c.statistics_id == \
    dic_table['statistics'].c.id)).order_by(dic_table['statistics'].c.min_played.desc()).limit(20)
mapper_results = engine.execute(mapper_stmt).fetchall()
most_plaing_players = pd.DataFrame(mapper_results, columns=('player_id','player_name','min'))
most_plaing_players
```
